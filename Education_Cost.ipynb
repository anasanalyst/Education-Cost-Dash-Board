```python
#Lets find the Average Fee of students based on top 10 countires without bringing the level of education 
Average_CountryFee=df.groupby("Country")["Tuition_USD"].mean().sort_values(ascending=False).nlargest(10).reset_index()
Average_CountryFee
```

```python
#Using Bargrapgh to visualize the data to understand it easily
plt.figure(figsize=(16,6))
plt.style.use("fivethirtyeight")
sns.set_context("paper", font_scale=1.5)
sns.barplot(Average_CountryFee, x="Country", y="Tuition_USD", palette="Set3")
```

```python
# Finding the Cost of living VS the Tuition Fee
df["Cost_Living"]=df["Rent_USD"]* 12 + df['Living_Cost_Index']* 100
df["Cost_Living"].reset_index().head(5)
```

```python
#Plotting a scatter plot to know the difference between Tuition-Fee and Cost of living
plt.figure(figsize=(16,10))
plt.style.use('bmh')
sns.set_context(font_scale=2)
sns.scatterplot(data=df, x="Tuition_USD", y="Living_Cost_Index", hue="Country",
               palette="tab20", s=100)
plt.title('Tuition Fees vs. Cost of Living by Country')
plt.xlabel('Tuition (USD)')
plt.ylabel('Estimated Annual Living Cost (USD)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```

```python
#Visualising the Fee Difference between the different level of education to have a much better understanding 
plt.figure(figsize=(16,8))
plt.style.use('fivethirtyeight')
sns.set_context("paper", font_scale=2)
sns.barplot(data=df, x='Level', y='Tuition_USD', palette="Set2")
plt.title('Tuition Fees Distribution by Degree Level (Log Scale)')
plt.xlabel('Degree Level')

plt.ylabel('Tuition USD')
plt.show()
```

```python
#Finding the University with highest tuition fee
Uni_Fee=df.groupby('University')['Tuition_USD'].max().sort_values(ascending=False).nlargest(10).reset_index()
Uni_Fee
```

```python
#With the Above finding we will Visualise by using barplot to understand the data
plt.figure(figsize=(14,6))
plt.style.use('bmh')
sns.set_context('paper', font_scale=1.5)
sns.barplot(Uni_Fee, x="University", y="Tuition_USD", palette='rocket')
plt.xticks(rotation=45,ha="right")
plt.xlabel('UNIVERSITIES')
plt.ylabel("Tuition-Fee")
plt.title("TOP 10 EXPENSIVE UNI'S")
plt.tight_layout()
```
